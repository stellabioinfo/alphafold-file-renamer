```python
import os

# Update this path to the directory containing AlphaFold output files
directory = r'/path/to/alphafold/files'

# Iterate through files in the directory
for filename in os.listdir(directory):
    # Check if the file name matches the AlphaFold pattern
    if filename.startswith('AF-') and filename.endswith('-F1-model_v4.pdb.gz'):
        # Extract the UniProt accession
        new_name = filename[3:-19] + '.pdb'

        # Rename file
        os.rename(
            os.path.join(directory, filename),
            os.path.join(directory, new_name)
        )
```
