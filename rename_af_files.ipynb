```python
# This script renames Alphafold files to only their Uniprot accession code

import os
import shutil
#import rename

# Set the directory where the files are located
directory = (r'\Users\\stell\\OneDrive\\Documentos\\tu_dresden\\ws22_23\\lab_project\\alphafold')

# Use a for loop to iterate over the files in the directory
for filename in os.listdir(directory):
    # Check if the file has the correct pattern
    if filename.startswith('AF-') and filename.endswith('-F1-model_v4.pdb.gz'):
        try:
            # Split the filename into parts
            parts = filename.split('-')
            # Create the new filename by combining the desired parts
            new_filename = parts[1]+'.pdb'
            print(new_filename)
            # Construct the full file path for the old and new filenames
            old_path = os.path.join(directory, filename)
            new_path = os.path.join(directory, new_filename)
            # Use the shutil module to rename the file
            shutil.move(old_path, new_path)
            #rename.move(old_path, new_path)
    elif not filename.startswith('AF-') and filename.endswith('-F1-model_v4.pdb'): filename = filename
```
