```python
"""
Rename AlphaFold structure files to use only the UniProt accession code.

This script expects files with the pattern:
    AF-<ACCESSION>-F1-model_v4.pdb.gz

It renames them to:
    <ACCESSION>.pdb
"""

import os
import shutil
from pathlib import Path

# Set the directory where your AlphaFold structure files are stored
# Replace this with the folder path on your machine before running
directory = Path(r'/path/to/alphafold/files')

# Iterate through the files in the directory
for filename in os.listdir(directory):

    # Check for the expected AlphaFold naming pattern
    if filename.startswith('AF-') and filename.endswith('-F1-model_v4.pdb.gz'):
        try:
            # Extract the UniProt accession from the filename
            parts = filename.split('-')
            new_filename = parts[1] + '.pdb'
            print(f"Renaming to: {new_filename}")

            # File paths
            old_path = directory / filename
            new_path = directory / new_filename

            # Rename (move) the file
            shutil.move(old_path, new_path)

        except Exception as e:
            print(f"Error renaming {filename}: {e}")

    # Secondary case: already-clean filenames
    elif filename.endswith('-F1-model_v4.pdb'):
        # No renaming needed, but kept for clarity
        pass
```
